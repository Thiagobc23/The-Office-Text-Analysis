```python
import pandas as pd
import numpy as np
from IPython.display import IFrame
```

```python
fields = ['name', 'text', 'episode_number', 'season', 'episode_name']
file = 'data/the_office_features.csv'
df =pd.read_csv(file, sep=';', encoding='utf-16')
```

```python
df.head()
```

```python
def search(name='', season=0, term='', name_contains=''):
    result = df.dropna()
    #CONDITIONS
    #name
    if(name != ''):
        result = result[result['name'].str.lower() == name.lower()]
    #season
    if (type(season) == list):
        result = result[result['season'].isin(season)]
    elif(season != 0):
        result = result[result['season'] == season]
    #term
    if(term != ''):
        result = result[result['text'].str.contains(term.lower(), regex=False, case=False)]
    #name contains
    if(name_contains != ''):
        result = result[result['name'].str.contains(name_contains.lower(), regex=False, case=False)]
    
    # PRINT RESULT SUMMARY
    print(str(result.shape[0]) + ' dialogs found.\n')
    
    print('With ' + str(len(result['name'].unique())) + ' unique character(s):')
    print(result['name'].unique(), '\n')
    
    print('And ' + str(len(result['ep_seas'].unique())) + ' episode(s) from ' +
          str(len(result['season'].unique())) + ' season(s)')
    return result
```

# Case Example:
## "That's what she said"

Michael Scott, played by Steve Carell is known for his "that's what she said" jokes.  
  
![Michael](https://upload.wikimedia.org/wikipedia/en/d/dc/MichaelScott.png)  
  
From https://en.wikipedia.org/wiki/Michael_Scott_(The_Office):

Legacy of "That's What She Said"
The show often uses the joke "that's what she said", which was popularized by the Wayne's World sketch on Saturday Night Live. In the original BBC version of The Office, Ricky Gervais's character David Brent frequently used the similar phrase "as the actress said to the bishop" as an inappropriate joke. Michael inserts the phrase as a sexually suggestive double entendre even in the most inappropriate circumstances, including business meetings and legal depositions. Michael finds uttering the phrase so irresistible that in "Sexual Harassment" he is goaded into saying it just seconds after Jan Levinson and a lawyer from Corporate specifically ask him not to do so. **The phrase has become so associated with the character that the television show 30 Rock in the episode "TGS Hates Women", there was a scene in which Liz Lemon became infuriated at another character's use of "TWSS" because "Steve Carell owns 'That's What She Said,' okay? He owns it!"** In the episode "Goodbye, Michael", "that's what she said" was Steve Carell's final (inaudible) line as a series regular, and was his first line upon returning as a guest star in "Finale".

## Catchphrase = Repetition ?  
  
![bugs bunny](https://media1.tenor.com/images/7bb00745e00863ea1f89fdcda1f0c758/tenor.gif?itemid=14258920)  
  
Most people would assume that a catchphrase is associated with the amount of times the phrase was used by the character, and since 'that's what she said' is Michael's catchphrase we would expect to find it many times trough the series.
  
With the collected data and a simple querying function we can see if this is true.

```python
pd.set_option('display.max_colwidth()', 100)

search(term="that's what she said", name="Michael")[fields]
```

## 23 dialogs in 20 episodes

This is not a big number, 20 episodes is only 10% of the series and in the vast majority of those 20 episodes michael only said the joke one time.

So how did they made this such a legacy to Steve Carell's character?  

To understand that we look at the beggining, when was the joke first used and how it was introduced.

```python
search(term="that's what she said", name="Michael", season=[1,2])[fields]
```

At the above table we can see the joke only came by at episode 2 of season 2.
